```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

df = pd.read_csv('seoultotal.csv', encoding = 'cp949')
df['bus'] = df['bus'] / 1000000
df['subway'] = df['subway'] / 1000000
df['taxi'] = df['taxi'] / 1000000
df['bicycle'] = df['bicycle'] / 1000000
df21 = df[df['year'] == 2021]
df22 = df[df['year'] == 2022]
df23 = df[df['year'] == 2023]
df24 = df[df['year'] == 2024]

df.head()
# downtown-도심권, northeast-동북권, northwest-서북권, southwest-서남권, southeast-동남권
```

```python
plt.figure(figsize = (20, 15))
p1 = plt.subplot(221)
p1.plot(df21['region'], df21['bus'], color='blue', linestyle=':', marker='o')
p1.plot(df21['region'], df21['subway'], color='green', linestyle=':', marker='o')
p1.plot(df21['region'], df21['taxi'], color='orange', linestyle=':', marker='o')
p1.plot(df21['region'], df21['bicycle'], color='red', linestyle=':', marker='o')
p1.set_xlabel('region')
p1.set_ylabel('sum_mill')
p1.set_ylim(0, 500)
p1.set_yticks(range(-50, 550, 50))
p1.set_title("2021")
p1.legend(['bus', 'subway', 'taxi', 'bicycle'])
p1.grid()

p2 = plt.subplot(222)
p2.plot(df22['region'], df22['bus'], color='blue', linestyle=':', marker='o')
p2.plot(df22['region'], df22['subway'], color='green', linestyle=':', marker='o')
p2.plot(df22['region'], df22['taxi'], color='orange', linestyle=':', marker='o')
p2.plot(df22['region'], df22['bicycle'], color='red', linestyle=':', marker='o')
p2.set_xlabel('region')
p2.set_ylabel('sum_mill')
p2.set_ylim(0, 500)
p2.set_yticks(range(-50, 550, 50))
p2.set_title("2022")
p2.legend(['bus', 'subway', 'taxi', 'bicycle'])
p2.grid()

plt.subplots_adjust(hspace = 0.2)

p3 = plt.subplot(223)
p3.plot(df23['region'], df23['bus'], color='blue', linestyle=':', marker='o')
p3.plot(df23['region'], df23['subway'], color='green', linestyle=':', marker='o')
p3.plot(df23['region'], df23['taxi'], color='orange', linestyle=':', marker='o')
p3.plot(df23['region'], df23['bicycle'], color='red', linestyle=':', marker='o')
p3.set_xlabel('region')
p3.set_ylabel('sum_mill')
p3.set_ylim(0, 500)
p3.set_yticks(range(-50, 550, 50))
p3.set_title("2023")
p3.legend(['bus', 'subway', 'taxi', 'bicycle'])
p3.grid()

p4 = plt.subplot(224)
p4.plot(df24['region'], df24['bus'], color='blue', linestyle=':', marker='o')
p4.plot(df24['region'], df24['subway'], color='green', linestyle=':', marker='o')
p4.plot(df24['region'], df24['taxi'], color='orange', linestyle=':', marker='o')
p4.plot(df24['region'], df24['bicycle'], color='red', linestyle=':', marker='o')
p4.set_xlabel('region')
p4.set_ylabel('sum_mill')
p4.set_ylim(0, 500)
p4.set_yticks(range(-50, 550, 50))
p4.set_title("2024")
p4.legend(['bus', 'subway', 'taxi', 'bicycle'])
p4.grid()
```
