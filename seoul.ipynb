```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

df = pd.read_csv('stotal.csv', encoding = 'cp949')
df['버스'] = df['버스'] / 1000000
df['지하철'] = df['지하철'] / 1000000
df['택시'] = df['택시'] / 1000000
df['자전거'] = df['자전거'] / 1000000
df21 = df[df['년도'] == 2021]
df22 = df[df['년도'] == 2022]
df23 = df[df['년도'] == 2023]
df24 = df[df['년도'] == 2024]

df.head()
# downtown-도심권, northeast-동북권, northwest-서북권, southwest-서남권, southeast-동남권
```

```python
plt.figure(figsize = (20, 15))
p1 = plt.subplot(221)
p1.plot(df21['구분'], df21['버스'], color='blue', linestyle=':', marker='o')
p1.plot(df21['구분'], df21['지하철'], color='green', linestyle=':', marker='o')
p1.plot(df21['구분'], df21['택시'], color='orange', linestyle=':', marker='o')
p1.plot(df21['구분'], df21['자전거'], color='red', linestyle=':', marker='o')
p1.set_xlabel('region')
p1.set_ylabel('sum_mill')
p1.set_ylim(0, 500)
p1.set_yticks(range(-50, 550, 50))
p1.set_title("2021")
p1.legend(['bus', 'subway', 'taxi', 'bicycle'])
p1.grid()

p2 = plt.subplot(222)
p2.plot(df22['구분'], df22['버스'], color='blue', linestyle=':', marker='o')
p2.plot(df22['구분'], df22['지하철'], color='green', linestyle=':', marker='o')
p2.plot(df22['구분'], df22['택시'], color='orange', linestyle=':', marker='o')
p2.plot(df22['구분'], df22['자전거'], color='red', linestyle=':', marker='o')
p2.set_xlabel('region')
p2.set_ylabel('sum_mill')
p2.set_ylim(0, 500)
p2.set_yticks(range(-50, 550, 50))
p2.set_title("2022")
p2.legend(['bus', 'subway', 'taxi', 'bicycle'])
p2.grid()

plt.subplots_adjust(hspace = 0.2)

p3 = plt.subplot(223)
p3.plot(df23['구분'], df23['버스'], color='blue', linestyle=':', marker='o')
p3.plot(df23['구분'], df23['지하철'], color='green', linestyle=':', marker='o')
p3.plot(df23['구분'], df23['택시'], color='orange', linestyle=':', marker='o')
p3.plot(df23['구분'], df23['자전거'], color='red', linestyle=':', marker='o')
p3.set_xlabel('region')
p3.set_ylabel('sum_mill')
p3.set_ylim(0, 500)
p3.set_yticks(range(-50, 550, 50))
p3.set_title("2023")
p3.legend(['bus', 'subway', 'taxi', 'bicycle'])
p3.grid()

p4 = plt.subplot(224)
p4.plot(df24['구분'], df24['버스'], color='blue', linestyle=':', marker='o')
p4.plot(df24['구분'], df24['지하철'], color='green', linestyle=':', marker='o')
p4.plot(df24['구분'], df24['택시'], color='orange', linestyle=':', marker='o')
p4.plot(df24['구분'], df24['자전거'], color='red', linestyle=':', marker='o')
p4.set_xlabel('region')
p4.set_ylabel('sum_mill')
p4.set_ylim(0, 500)
p4.set_yticks(range(-50, 550, 50))
p4.set_title("2024")
p4.legend(['bus', 'subway', 'taxi', 'bicycle'])
p4.grid()
```
